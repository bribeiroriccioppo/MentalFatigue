## Eyes open vs Eyes closed

```python
for name_id in range(len(participants_list)):
    chs_powers = []
    acc_powers = []
    blocks = []
    bands = []
    x_block_all = []
    for band, _, _ in data.FREQ_BANDS: # iteration in each frequency band
        if band == 'Alpha1' or band == 'Alpha2':
            for index, row in powers_df[(powers_df.participant == participants_list[name_id]) & (powers_df.band == band)].iterrows(): # powers
                if not np.isnan(row['All_rel']).any():
                    block = row['block']
                    chs_powers = list(row['All_rel'])
                    acc_powers += chs_powers
                    blocks += [block] * len(chs_powers)
                    bands += [band] * len(chs_powers)

    # Create dataframe
    df = pd.DataFrame([acc_powers, blocks, bands], index=['power', 'block', 'band']).T
    df = df[df.block.isin([1, 2, 3])]
    
    power_alpha1 = df.loc[df['band'] == 'Alpha1']['power']
    #power_alpha2 = df.loc[df['band'] == 'Alpha2']['power']
    x = range(len(power_alpha1))
    nb_blocks = 3
    x_block = 0
    for i in range(nb_blocks):
        df_band = df.loc[df['band'] == 'Alpha1']
        df_block = df_band.loc[df_band['block'] == i+1]
        x_block += len(df_block)
        x_block_all.append(x_block)
    
    # Plots
    plt.figure()
    plt.xlabel("")
    plt.ylabel("Relative power")
    plt.plot(x,power_alpha1,label="Alpha 1")
    # SavitzkyGolay filter: digital filter to smooth the data by fitting successive small windows of data with a low-degree polynomial by the method of linear least squares
    smooth_power_alpha1 = savgol_filter(power_alpha1, 51, 3) # window size 51, polynomial order 3
    plt.plot(smooth_power_alpha1)
    #plt.plot(x,power_alpha2,label="Alpha 2")
    plt.axvline(x_block_all[0], 0, 1, label='End Eyes Open', color='c')
    plt.axvline(x_block_all[1], 0, 1, label='Start Eyes Closed', color='g')
    plt.axvline(x_block_all[2], 0, 1, label='End Eyes Closed', color='r')
    plt.legend()
```
