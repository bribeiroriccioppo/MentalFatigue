### Pupil size evolution per block

```python
blocks = tobii_df.block_idx.unique()
fig, axs = plt.subplots(len(blocks), 1, figsize=(20, 50), sharey=True)
for idx, block_idx in enumerate(blocks):
    tobii_df[tobii_df.block_idx == block_idx].plot.scatter(x='lsl_ts', y='left_pupil', ax=axs[idx], color='blue')
    tobii_df[tobii_df.block_idx == block_idx].plot.scatter(x='lsl_ts', y='right_pupil', ax=axs[idx], color='orange')
    axs[idx].set_title("Block {}".format(int(block_idx)))
    axs[idx].set_xlabel("Time")
    axs[idx].set_ylabel("Pupil (mm)")
```

### Pupil dilation evolution per block

```python
blocks = tobii_df.block_idx.unique()
fig, axs = plt.subplots(len(blocks), 1, figsize=(20, 50), sharey=True)
for idx, block_idx in enumerate(blocks):
    tobii_df[tobii_df.block_idx == block_idx].plot.scatter(x='lsl_ts', y='left_pupil_pct', ax=axs[idx], color='blue')
    tobii_df[tobii_df.block_idx == block_idx].plot.scatter(x='lsl_ts', y='right_pupil_pct', ax=axs[idx], color='orange')
    axs[idx].set_title("Block {}".format(int(block_idx)))
    axs[idx].set_xlabel("Time")
    axs[idx].set_ylabel("Pupil (mm)")
```

### Pupil dilation evolution during dual task

#### Block 1

```python
# 13, 15, 17 and 19
tasks_block_1 = [13.0, 15.0, 17.0, 19.0]
fig, ax = plt.subplots(figsize=(12,4))

for idx, block_idx in enumerate(tasks_block_1):
    sns.boxplot(x='block_idx', y='pupil_pct', data=tobii_df[tobii_df.block_idx.isin(tasks_block_1)], ax=ax)
    ax.set_xlabel("Condition")
    ax.set_ylabel("Pupil Dilation (%)");
```
